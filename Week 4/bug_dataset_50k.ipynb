```python
import pandas as pd
import numpy as np

# Load Dataset
df = pd.read_csv(r"D:\internship as data analyst\Codiora softwatre house\Week 4\bug_dataset_50k.csv")

# ==========================
# Explore Dataset
# ==========================
print("First 5 Rows:")
print(df.head())

print("\nDataset Information:")
print(df.info())

print("\nDataset Shape:")
print(df.shape)

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())

# ==========================
# Data Cleaning
# ==========================

# Remove duplicate rows
df = df.drop_duplicates()

# Remove duplicate bug IDs
df = df.drop_duplicates(subset="bug_id")

# Fill missing values in error_code with median
df["error_code"] = df["error_code"].fillna(df["error_code"].median())

# Remove leading/trailing spaces from text columns
text_cols = df.select_dtypes(include="object").columns
df[text_cols] = df[text_cols].apply(lambda x: x.str.strip())

# ==========================
# Standardize Formats
# ==========================

# Convert created_at to datetime
df["created_at"] = pd.to_datetime(df["created_at"])

# Convert selected text columns to Title Case
columns = [
    "bug_category",
    "bug_domain",
    "tech_stack",
    "severity",
    "environment",
    "developer_role",
    "root_cause"
]

for col in columns:
    df[col] = df[col].str.title()

# ==========================
# Create Calculated Fields
# ==========================

# Year
df["Year"] = df["created_at"].dt.year

# Month
df["Month"] = df["created_at"].dt.month_name()

# Quarter
df["Quarter"] = df["created_at"].dt.quarter

# Day of Week
df["Day"] = df["created_at"].dt.day_name()

# High Severity Flag
df["High_Severity"] = np.where(
    df["severity"] == "Critical",
    "Yes",
    "No"
)

# Error Code Available
df["Error_Code_Available"] = np.where(
    df["error_code"] > 0,
    "Yes",
    "No"
)

# ==========================
# Final Validation
# ==========================

print("\nFinal Dataset Shape:")
print(df.shape)

print("\nMissing Values After Cleaning:")
print(df.isnull().sum())

print("\nDuplicate Rows After Cleaning:")
print(df.duplicated().sum())

print("\nData Types:")
print(df.dtypes)

print("\nFirst 5 Rows After Cleaning:")
print(df.head())

# ==========================
# Save Cleaned Dataset
# ==========================

df.to_csv(
    r"D:\internship as data analyst\Codiora softwatre house\Week 4\Cleaned_Bug_Dataset.csv",
    index=False
)

print("\nData cleaning completed successfully!")
print("Cleaned dataset saved as 'Cleaned_Bug_Dataset.csv'")
```
